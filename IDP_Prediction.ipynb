```python
X_train = pd.concat([X_train,y_train],axis=1)
```

```python
ipd.rename(columns={'Disease Label': 'Disease_Label'}, inplace=True)
X_train.rename(columns={'Disease Label': 'Disease_Label'}, inplace=True)
```

```python
X_train_IPD = X_train.query('Disease_Label == 1')
X_train_Control = X_train.query('Disease_Label == 0')
print(X_train_IPD)
print(X_train_Control)
```

```python
X_train_IPD.dtypes
```

```python
from scipy import stats
signif_genes={}
for gene in X_train.columns[:-1]:
  _,p = stats.mannwhitneyu(X_train_IPD[gene], X_train_Control[gene])
  if p<0.05:
    print("__")
    signif_genes[gene] = p
```

```python
print(signif_genes)
```

```python
len(signif_genes)
```

```python
genes = []
p_values = []
items = signif_genes.items()
for item in items:
    genes.append(item[0]), p_values.append(item[1])
```

```python
print(genes)
print(p_values)
```

```python
from statsmodels.stats import multitest

is_null_hypothesis_rejected, p_values_corrected = multitest.fdrcorrection(p_values, alpha=0.05, method='indep', is_sorted=False)
```

```python
print(p_values_corrected)
```

```python
len(p_values_corrected)
```

```python
print(is_null_hypothesis_rejected)
```

```python
np.count_nonzero(is_null_hypothesis_rejected == True)
```

```python
is_null_hypothesis_rejected, p_values_corrected, _, __ = multitest.multipletests(p_values, alpha=0.05, method='holm-sidak', is_sorted=False, returnsorted=False)
```

```python
np.count_nonzero(is_null_hypothesis_rejected == True)
```

```python
is_gene_significant = dict(zip(genes, is_null_hypothesis_rejected))
```

```python
for gene in genes:
  if is_gene_significant[gene]==True:
    print(gene)
```

```python
final_list_of_genes=[]

for gene in is_gene_significant:
  if is_gene_significant[gene]==True:
    final_list_of_genes.append(gene)
```

```python
X_train=X_train[final_list_of_genes]
X_test=X_test[final_list_of_genes]
```

```python
import pandas as pd
import numpy as np

def log2_normalize(df):
    epsilon = 1e-8

    # Exclude the first column from normalization
    excluded_column = df.iloc[:, 0]

    # Apply log2 transformation to the remaining columns
    log2_data = np.log2(df.iloc[:, 1:] + epsilon)

    # Add back the excluded column
    normalized_data = pd.concat([excluded_column, log2_data], axis=1)

    return normalized_data

# Assuming you have a gene expression dataframe named 'gene_expression_df'
# where the first column represents sample names and the remaining columns represent genes

# Perform log2 normalization excluding the first column
normalized_df = log2_normalize(ipd)

# Print the normalized dataframe
print(normalized_df)
```
